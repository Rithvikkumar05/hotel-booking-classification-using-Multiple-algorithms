```python
df = df.dropna()
```

```python
df = df.drop(['agent','credit_card','name','adr','email','phone-number'],axis =1)
```

```python
X = df.drop('is_canceled',axis =1)
Y = df['is_canceled']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size =.2)
```

```python
from sklearn.linear_model import LogisticRegression
xdd  = LogisticRegression()
xdd.fit(x_train,y_train)
```

```python
y_p1 = xdd.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_p1,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf1 = DecisionTreeClassifier(criterion = 'entropy')
clf1.fit(x_train,y_train)
```

```python
y_p2 = clf1.predict(x_test)
```

```python
accuracy_score(y_p2,y_test)
```

```python
x_pred = clf1.predict(x_train)
accuracy_score(x_pred,y_train)
```

```python
from sklearn.svm import SVC
clf2 = SVC(kernel = 'rbf')
clf2.fit(x_train,y_train)
```

```python
y_p3 = clf2.predict(x_test)
```

```python
accuracy_score(y_p3,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf3 = RandomForestClassifier(n_estimators =150,criterion='log_loss',max_features = 10 )
clf3.fit(x_train,y_train)
```

```python
y_P4 = clf3.predict(x_test)
```

```python
accuracy_score(y_P4,y_test)
```
